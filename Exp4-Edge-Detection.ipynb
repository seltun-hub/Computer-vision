# Experiment 4: Edge Detection
Design and implement a Python program using OpenCV to perform edge detection based on the
Canny algorithm.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

# Load image
!wget -q -O input.jpg https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg
image = cv2.imread('input.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Canny Edge Detection
# Syntax: cv2.Canny(image, threshold1, threshold2)
# threshold1 = minVal, threshold2 = maxVal
edges_wide = cv2.Canny(gray, 50, 150)
edges_tight = cv2.Canny(gray, 180, 240)

# Visualization
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(edges_wide, cmap='gray')
plt.title('Wide Threshold (50, 150)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edges_tight, cmap='gray')
plt.title('Tight Threshold (180, 240)')
plt.axis('off')

plt.show()
```
