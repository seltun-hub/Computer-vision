# Experiment 3: Histogram Analysis
Write a program to compute and plot the histogram of a grayscale image and perform histogram
equalization.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os

# Download image
!wget -q -O input_image.jpg https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg

# Load image
img = cv2.imread('input_image.jpg', cv2.IMREAD_GRAYSCALE)


# Perform Histogram Equalization
img_equalized = cv2.equalizeHist(img)

# Create the visualization plot
plt.figure(figsize=(16, 8))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Original Histogram
plt.subplot(2, 2, 2)
plt.hist(img.ravel(), 256, [0, 256], color='black')
plt.title('Original Histogram')
plt.xlim([0, 256])

# Equalized Image
plt.subplot(2, 2, 3)
plt.imshow(img_equalized, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

# Equalized Histogram
plt.subplot(2, 2, 4)
plt.hist(img_equalized.ravel(), 256, [0, 256], color='black')
plt.title('Equalized Histogram')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
