# Experiment 7: Corner Detection
Design and implement a corner detection algorithm using Harris / SIFT Techniques

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

# Load image
!wget -q -O input.jpg https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg
image = cv2.imread('input.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Harris Corner Detection
# Parameters: (image, blockSize, ksize, k)
# gray must be float32
gray_float = np.float32(gray)
dst = cv2.cornerHarris(gray_float, 2, 3, 0.04)

# Result
dst = cv2.dilate(dst, None)

# Threshold for an optimal value
harris_img = image.copy()
harris_img[dst > 0.01 * dst.max()] = [0, 0, 255] # Mark corners in Red

print("Harris Corner Detection:")
cv2_imshow(harris_img)


# Initialize SIFT detector
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors
keypoints, descriptors = sift.detectAndCompute(gray, None)

# Draw keypoints on the image
sift_img = cv2.drawKeypoints(image, keypoints, None, 
                             flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

print("SIFT Keypoint Detection:")
cv2_imshow(sift_img)
```
